## Задание 5

Нахождение сигнала, проходящего через согласованный фильтр с помощью комплексной огибающей $c_x$

<img src="/m4L4qIUKeq0.jpg" alt="board" width="750"/>



```python
import numpy as np
import scipy.fft as fft 
import scipy.signal as signal
import matplotlib.pyplot as plt
from math import *

delta_w = 50
T = 1

h = 0.0001
t_len = int(T/h)
ts = np.linspace(0,T,t_len)

c_x = np.exp((0 + 1j) * ( - delta_w * ts + 2.0 / T * delta_w * ts *ts))

c_x_real = np.real(c_x)
c_x_imag = np.imag(c_x)
```

Действительная и мнимая часть комплексной огиб сигнала x

```python
fig = plt.figure(figsize=(16,8))
plt.plot(c_x_real)
plt.plot(c_x_imag)
```

```python
c_h = np.exp((0 - 1j) * ( - delta_w * (T - ts) + 2.0 / T * delta_w * (T - ts) * (T - ts)))
c_h_real = np.real(c_h)
c_h_imag = np.imag(c_h)
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot(c_h_real)
plt.plot(c_h_imag)
```

```python
c_y = 0.5  * signal.convolve(c_h, c_x)
c_y_real = np.real(c_y)
c_y_imag = np.imag(c_y)
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot(c_y_real, label="real part c_y")
plt.plot(c_y_imag, label="image part c_y")
plt.legend()
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot(np.abs(c_y), label="Abs c_y")
plt.legend()
plt.title("c_y(t) Через свертку c_x и с_h")
```

```python
fig = plt.figure(figsize=(16,8))

c_x_1 = np.hstack((c_x, np.zeros(t_len)))
c_h_1 = np.hstack((c_h, np.zeros(t_len)))

c_x_2 = fft.fft(c_x_1)
c_h_2 = fft.fft(c_h_1)
c_y_fft = 0.5* c_x_2 *c_h_2
c_y_new =  fft.ifft(c_y_fft)

plt.plot(np.abs(c_y_new))
plt.title("Abs(C_y) через преобразование фурье");
```

```python
# Получение выходного сигнала по формуле
fig = plt.figure(figsize=(16,8))
delta = T / (t_len)
ts = np.arange(0, 2*T, delta)
w0 = 1000
y = np.real(c_y_new * np.exp((0 + 1j) * w0 * ts))
plt.plot(y)
```
