```python
# Задача 3 ЛЧМ фильтр

import numpy as np
import scipy.fft as fft 
import scipy.signal as signal
import matplotlib.pyplot as plt
from math import *

Omega = lambda t, w_min, alpha : w_min + alpha * t
X = lambda t, w_min, alpha, phi=0 : np.cos(Omega(t, w_min, alpha) * t + phi)
```

```python
#Получение частотной характеристики ЛЧМ сигнала

w_min = 1000
w_max = 10000
t0 = 0
t1 = 1
h = 0.01
freq = lambda t: w_min+alpha*t
t_len = int((t1-t0)/h)
alpha = (w_max-w_min)/t_len
T=np.linspace(t0,t1,t_len)
x = np.asarray([cos(t*freq(t)) for t in T])# добавить нули
x_new = np.hstack((x, np.zeros(20*t_len)))
y = np.fft.fft(x_new)
plt.plot(np.abs(y))
y_1 = y[:int(len(y)/2)]
fig = plt.figure(figsize=(16,8))
plt.plot(abs(y_1), color = 'blue')
```

```python
X_prod = x_new * x_new
res = fft.ifft(X_prod)
plt.plot(np.abs(res))
N_res = len(res)
```

```python
res_res = np.hstack((res[N_res//2:N_res], res[0:N_res//2]))
```
