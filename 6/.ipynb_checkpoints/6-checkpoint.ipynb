Задание:

Дан ЛЧМ сигнал (увеличивающаяся частота) с переменной амплитудой A(t) и переменной w(t)
мы хотим по этому сигналу найти характеристики
1. A(t) амлитуду
2. PHI(t) 
3. f(t) частоту

Делается это в три приёма:
1. C помощью преобразования фурье находим X(f)
2. В спетре зануляем правую половину и увеличиваем в два раза оставшуюся часть -- это и есть спектр
аналитического сигнала z(t)
3. Находим характеристики по формулам с доски


Что МОЖНО менять для эксперимента: 
1. функции амплитуды и w(t) на произвольные
2. число отсчетов N, длину сигнала T, константу phi_0

```python
import numpy as np
import scipy.fft as fft 
import scipy.signal as signal
import matplotlib.pyplot as plt

A = lambda t : np.sin(t/np.pi)
w_0  = lambda t : 5.0*t + 3
phi_0  = 1000
T = 10
N = 1000
ts = np.linspace(0, T, N)
fig = plt.figure(figsize=(16,10))
x = A(ts) * np.cos( w_0(ts) * ts + phi_0)
plt.plot(ts, x)

Φ = w_0(ts)*ts + phi_0

```

```python
fig = plt.figure(figsize=(16,10))
x_fft = fft.fft(x)

x_fft[int(N/2):] = np.zeros(int(N/2))
x_fft[:int(N/2)] = x_fft[:int(N/2)] * 2

fig = plt.figure(figsize=(16,10))
plt.plot(np.abs(x_fft))


```

Если мы все сделали правильно, то действительная часть от z(t) должна совпасть с исходным сигналом x(t)

```python
z = fft.ifft(x_fft)
plt.plot(np.real(z))
```

```python
fig = plt.figure(figsize=(16,10))
A_done = np.abs(z)
plt.plot(A_done)
plt.plot(A(ts))
plt.title("Сравнение найденной амплитуды  задаваемой ")
```

```python
fig = plt.figure(figsize=(16,10))
PHI = np.angle(z)
plt.plot(PHI)
PHI_unwrap = np.unwrap(PHI)
plt.plot()
```

P.S большая PHI получается с дикими разрывами (см правый низ доски), поэтому пользуемся функцией unwrap 
Это нужно для того, чтобы корректно найти производную в формуле для f(t)

```python
fig = plt.figure(figsize=(16,10))
phi = 1.0 / 2.0 / np.pi * np.gradient(PHI_unwrap)
plt.plot(phi)
plt.title("Увеличивающаяся частота сигнала")
mean2 = 2*np.pi/T * np.mean(phi)
mean3 = 2*np.pi/T * np.trapz(phi, ts)
mean2, mean3
```
