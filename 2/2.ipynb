```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
import scipy.signal as s
from math import sin, cos, exp, sqrt, cosh, log, pi
import math 

import statsmodels.api as sm
lowess = sm.nonparametric.lowess
```

На вход цифровой сигнал, подаётся шум, сигнал проходит через согласованный фильтр
передаточная функция $h(t) = X(T - t).$

Поэтому найти выходной сигнал после прохождения фильтра легко найти с помощью свертки $x*x$, где
$*$ - операция свертки (convolve) 

В конце работы сравниваются свертки (выходной сигнал), если бы шума не было, и с шумом

```python
n = 100
X = np.transpose([*np.zeros(n), *np.ones(n),*np.zeros(n)])
plt.plot(X)
```

```python
N = 0.5
F_max = 0.01 # n=100 F_max = 1/T = 0.01
sigma = math.sqrt(3 * N * F_max) * 20
S = np.random.standard_normal(3*n)

X_new = X + sigma*S

fig = plt.figure(figsize=(20,10))
plt.plot(X)
plt.plot(X_new);
```

```python
y_1 = s.convolve(X, X)
```

```python
y_2 = s.convolve(X, X_new)
fig = plt.figure(figsize=(20,12))
plt.plot(y_1)
plt.plot(y_2)
```
